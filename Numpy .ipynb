```python
n = int(input('Satır Sayısını Giriniz: '))
m = int(input('Sütun Sayısını Giriniz: '))
A = [] # Boş Liste Tanımladık 
for indexSatir in range(n):
  A_satir = [] # Elemanları Liste Olan Bir Liste Tanımlamalıyız
  for indexSutun in range(m):
    i = indexSatir + 1
    j = indexSutun + 1    
    A_satir.append(i ** 2 + 3 * j)
  A.append(A_satir)
A
```

```python
m = int(input('Satır Sayısının giriniz: '))
n = int(input('Sütun Sayısının giriniz: '))
A = np.zeros((m, n), int) # m satır n sütundan oluşan tam sayı tutan ve elemanları 0 olan 2 boyutlu numpy array tanımladık 
for i in range(1, m + 1):
  for j in range(1, n + 1):
    A[i - 1, j - 1] = i ** 2 + 3 * j
A
```

```python
m = int(input('Satır Sayısının giriniz: '))
n = int(input('Sütun Sayısının giriniz: '))
B = np.zeros((m, n), int)
for i in range(1, m + 1):
  for j in range(1, n + 1):
    if i == j:
      B[i - 1, j - 1] = i + j
    else:
      B[i - 1, j - 1] = i ** 2 + j
B
```

```python
def B_matris_bul(m, n):
  B = np.zeros((m, n), int)
  for i in range(1, m + 1):
    for j in range(1, n + 1):
      if i == j:
        B[i - 1, j - 1] = i + j
      else:
       B[i - 1, j - 1] = i ** 2 + j
  return B
```

```python
B_matris_bul(3, 2)
```

```python
def M_matris_bul(m, n):
  M = np.zeros((m, n), int)
  for i in range(1, m + 1):
    for j in range(1, n + 1):
      if j - i == 1:
        M [i - 1, j - 1] = i
  return M  
```

```python
M_matris_bul(5, 5)
```

```python
M_matris_bul(7, 9)
```

```python
def N_harfi_olustur(n):
  N = np.zeros((n, n), int)
  for i in range(1, n + 1):
    for j in range(1, n + 1):
      if   i == j or j == 1 or j == n:
        N[i - 1, j - 1] = 1
  return N 
```

```python
N_harfi_olustur(5)
```

```python
N_harfi_olustur(10)
```

```python
def X_harfi_olustur(n):
  X = np.zeros((n, n), int)
  for i in range(1, n + 1):
    for j in range(1, n + 1):
      if  i == j or i + j == n + 1:
        X[i - 1, j - 1] = 1
  return X 
```

```python
X_harfi_olustur(10)
```

```python
def Z_harfi_olustur(n):
  Z = np.zeros((n, n), int)
  for i in range(1, n + 1):
    for j in range(1, n + 1):
      if  i + j == n + 1 or i == 1 or i == n:
        Z[i - 1, j - 1] = 1
  return Z 
```

```python
Z_harfi_olustur(5)
```

```python
Z_harfi_olustur(10)
```

```python
def sonsoz_olustur(n):
  S = np.zeros((n, n), int)
  for i in range(1, n + 1):
    for j in range(1, n + 1):
      if  i + j == n + 1 or i == 1 or i == n or i == j:
        S[i - 1, j - 1] = 1
  return S 
```
